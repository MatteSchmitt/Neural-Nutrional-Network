```python
import pandas as pd
import numpy as np

# Load the DataFrame from the CSV file
df = pd.read_csv("food_results.csv")

# Define a function to extract energy (KCAL) and carbohydrate values using regular expressions
def extract_energy_and_carbs(details):
    import re
    if isinstance(details, str):
        energy_pattern = r"'nutrientName': 'Energy', 'nutrientNumber': '208', 'unitName': 'KCAL', 'value': (\d+)"
        carb_pattern = r"'nutrientName': 'Carbohydrate, by difference', 'nutrientNumber': '205', 'unitName': 'G', 'value': (\d+\.\d+)"
        
        energy_match = re.search(energy_pattern, details)
        carb_match = re.search(carb_pattern, details)
        
        energy_value = energy_match.group(1) if energy_match else np.nan
        carb_value = carb_match.group(1) if carb_match else np.nan
        
        return energy_value, carb_value
    else:
        return np.nan, np.nan

# Apply the function to extract energy and carbohydrate values from the 'Details' column
df['Energy_Value'], df['Carbohydrate_Value'] = zip(*df['Details'].apply(extract_energy_and_carbs))

# Save the updated DataFrame to a new CSV file
df.to_csv("cleaned_food_details2.csv", index=False)

# Display the updated DataFrame
print(df)
```
